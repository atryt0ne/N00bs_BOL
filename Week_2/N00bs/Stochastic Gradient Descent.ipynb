Stochastic Gradient Descent(SGD) is an optimization technique. <br>
1. What is Optimization?<br>
The process of minimizing (or maximizing) any mathematical expression is called optimization.<br>
2. There are various BATCH methods for optimization, but they come with two major drawbacks.<br>A) Computing the cost and gradient for the entire training set can be very slow if working on a single machine if the dataset is too big to fit in the main memory.<br>B) In an online setting, batch optimization methods don't easily incorporate new data.<br>
3. SGD addresses these issues. It follows the negative gradient of the objective after seeing a single or a few training examples. It leads to fast convergence and overcomes the cost of back propogation over the full data set.<br>


SGD<br>
In SGD, a few samples are selected randomly instead of the whole data set in each iteration.<br>
SGD Algorithm -<br> 
For i in range (m):<br>
$\theta_j = \theta_j -\alpha(y'^i - y^i)x^i_{j}$<br>
In SGD, we find the gradient of the cost function of a single sample at each iteration.<br>
As only one sample is chosen at random for each iteration, the path taken to reach the minima is noisier. But that does not matter as we reach the minima in a significantly small amount of time.<br>
